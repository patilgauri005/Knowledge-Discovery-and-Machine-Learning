### 1. Import Libraries and Load the Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your specific dataset
data = pd.read_csv("background_noise_focus_dataset.csv")

# Display the first 5 records
print("First 5 records of the dataset:")
print(data.head())
```

### 2. Dataset Structure and Information

```python
# Display basic information about the dataset
print("\nDataset Info:")
data.info()

# Display the shape (rows, columns)
print("\nDataset Shape:", data.shape)
```

### 3. Data Types and Columns


```python
# Display the names of all columns
print("\nColumns in Dataset:")
print(data.columns)

# Display data types
print("\nData Types:")
print(data.dtypes)
```

### 4. Descriptive Statistical Analysis

```python
# Statistical summary for numerical columns
print("\nStatistical Summary:")
data.describe()
```

### 5. Missing Value Analysis

```python
# Check for missing values
print("\nMissing Values in Each Column:")
print(data.isnull().sum())

# Display percentage of missing values
print("\nPercentage of Missing Values:")
print((data.isnull().sum() / len(data)) * 100)
```

### 6. Univariate Analysis (Histograms & Boxplots)

```python
# Plot histograms for all numerical columns
data.hist(figsize=(12, 10), bins=15, color='skyblue', edgecolor='black')
plt.suptitle("Histograms of Numerical Variables")
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
# Plot boxplots to detect outliers
plt.figure(figsize=(12, 6))
sns.boxplot(data=data.select_dtypes(include=[np.number]))
plt.title("Boxplots of Numerical Columns")
plt.xticks(rotation=45)
plt.show()
```

### 7. Multivariate Analysis (Scatter Plot)

```python
# Scatter plot between Focus Duration and Focus Score
plt.figure(figsize=(8, 6))
plt.scatter(data['focus_duration_minutes'], data['perceived_focus_score'], alpha=0.5, color='teal')
plt.xlabel("Focus Duration (minutes)")
plt.ylabel("Perceived Focus Score")
plt.title("Scatter Plot: Focus Duration vs Focus Score")
plt.show()
```

### 8. Correlation Analysis and Heatmap

```python
# Compute correlation matrix
corr = data.corr(numeric_only=True)
print("\nCorrelation Matrix:")
print(corr)

# Correlation Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

### 9. Outlier Detection Using IQR

```python
# Define column for outlier detection
col = 'focus_duration_minutes'
Q1 = data[col].quantile(0.25)
Q3 = data[col].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = data[(data[col] < lower_bound) | (data[col] > upper_bound)]

print(f"\nOutlier Detection for {col}:")
print(f"Q1: {Q1}, Q3: {Q3}, IQR: {IQR}")
print(f"Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")
print(f"Number of outliers detected: {len(outliers)}")
```
